```python
import time

time_yf = np.array([])

for i in range(50):
    start_time = time.time()

    data_yf = yf.download('AAPL', start='2015-6-24', end='2020-12-31')['Close']

    time_yf = np.append(time_yf, time.time() - start_time) 
```

```python
import time

time_eetc = np.array([])

for i in range(50):
    start_time = time.time()

    data_eetc = client.get_price_data('AAPL', from_date='2010-1-1', to_date='2020-12-31')['close']

    time_eetc = np.append(time_eetc, time.time() - start_time) 
```

```python
df_eetc = pd.DataFrame(time_eetc, columns=['EETC_Seconds']).rename_axis('Trial')
df_eetc.index+=1
df_eetc
```

```python
time = pd.concat([df_yf, df_eetc], axis=1)
time
```

```python
time.describe()
```

```python
plt.figure(figsize=(12,8))
sns.histplot(time['YF_Seconds'], color='blue')
sns.histplot(time['EETC_Seconds'], color='red', alpha=0.3)
plt.legend(['YF_Seconds', 'EETC_Seconds'])
plt.xlabel('Seconds')
plt.show()
```
